### Perform A/B testing to find the best campaign for the company to get more customers.

#### Install packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
```

#### Data Praperation

```python
control_data = pd.read_csv('control_group.csv', delimiter=';')
control_data.head()
```

```python
test_data = pd.read_csv('test_group.csv', delimiter=';')
test_data.head()
```

### AB Testing

Click-Through Rate (CTR): This is the ratio of users who click on an ad to the number of total users who view the ad (impressions). It gives an idea of how effective the ad is at capturing users' attention.

Conversion Rate: This is the ratio of users who take a desired action (e.g., making a purchase) to the total number of users who clicked on the ad.

Cost Per Acquisition (CPA): This is the average cost the company pays for each acquisition (or purchase in this case).

Return On Advertising Spend (ROAS): This metric provides an overview of the effectiveness of the advertising campaign. 

```python
# Calculating metrics for the control group
control_data['CTR'] = control_data['# of Website Clicks'] / control_data['# of Impressions']
control_data['Conversion Rate'] = control_data['# of Purchase'] / control_data['# of Website Clicks']
control_data['CPA'] = control_data['Spend [USD]'] / control_data['# of Purchase']

control_metrics = control_data[['CTR', 'Conversion Rate', 'CPA']].mean()
control_metrics
```

```python
# Calculating metrics for the test group
test_data['CTR'] = test_data['# of Website Clicks'] / test_data['# of Impressions']
test_data['Conversion Rate'] = test_data['# of Purchase'] / test_data['# of Website Clicks']
test_data['CPA'] = test_data['Spend [USD]'] / test_data['# of Purchase']

test_metrics = test_data[['CTR', 'Conversion Rate', 'CPA']].mean()
test_metrics
```

```python
# Data for plotting
labels = ['CTR', 'Conversion Rate', 'CPA']
control_values = control_metrics.values
test_values = test_metrics.values

x = range(len(labels))

# Creating the bar plot
plt.figure(figsize=(10, 6))
plt.bar(x, control_values, width=0.4, label='Control Group', align='center')
plt.bar(x, test_values, width=0.4, label='Test Group', bottom=control_values, align='center')

# Adding labels and title
plt.xlabel('Metrics')
plt.ylabel('Value')
plt.title('Comparison of Metrics between Control and Test Groups')
plt.xticks(x, labels)
plt.legend()
plt.grid(axis='y')

# Display the plot
plt.tight_layout()
plt.show()
```

The CTR is higher in the test group, indicating that the ad in the test group captures users' attention better.
However, the Conversion Rate is slightly higher in the control group, meaning that the proportion of users who actually make a purchase after clicking the ad is higher in the control group.
The CPA is also higher in the test group, implying that the company is paying more for each acquisition in the test group compared to the control group.

To determine the difference, I decided to do other statistical tests. 

#### Mann-Whitney U test

```python
results = {}

for label in labels:
    control_metric = control_data[label].dropna()  # Removing NaN values
    test_metric = test_data[label].dropna()  # Removing NaN values

    # Mann-Whitney U test
    u_statistic, p_value = mannwhitneyu(control_metric, test_metric, alternative='two-sided')
    results[label] = {
        'U Statistic': u_statistic,
        'P-Value': p_value
    }

results
```

The CTR of the test group is significantly different from the control group. There's no significant difference in the Conversion Rate and CPA between the two groups since the P-Values are greater than 0.05.

In conclusion, the test campaign has a significantly different CTR compared to the control campaign, but the conversion rate and CPA do not show significant differences between the two groups.

```python
# Extracting p-values for plotting
p_values = [results[label]['P-Value'] for label in labels]

# Creating the bar plot for p-values
plt.figure(figsize=(10, 6))
plt.bar(labels, p_values, color='grey', alpha=0.7)
plt.axhline(y=0.05, color='red', linestyle='--', label='Significance Threshold (p=0.05)')

# Adding labels, title, and legend
plt.xlabel('Metrics')
plt.ylabel('P-Value')
plt.title('P-Values from Mann-Whitney U Test for Each Metric')
plt.legend()
plt.ylim(0, 1)
plt.grid(axis='y')

# Display the plot
plt.tight_layout()
plt.show()
```

Metrics with blue bars below the red dashed line are considered to have statistically significant differences between the control and test groups. As observed, only the Click-Through Rate (CTR) shows a significant difference, while the Conversion Rate and Cost Per Acquisition (CPA) do not.

#### Conclusion

The test campaign successfully captures more attention and entices users to click on the advertisement.
However, once users click on the advertisement, their likelihood to make a purchase is similar for both campaigns.
The costs associated with acquiring a customer are also similar for both campaigns.

### User Funnel Analysis

```python
# Aggregating data for the funnel stages
funnel_stages = ['# of Impressions', '# of Website Clicks', '# of Searches', '# of View Content', '# of Add to Cart', '# of Purchase']
control_funnel_values = control_data[funnel_stages].sum()
test_funnel_values = test_data[funnel_stages].sum()

# Plotting the user funnel
plt.figure(figsize=(12, 8))
plt.plot(funnel_stages, control_funnel_values, marker='o', label='Control Group', color='blue')
plt.plot(funnel_stages, test_funnel_values, marker='o', label='Test Group', color='green')

# Adding labels, title, and legend
plt.xlabel('Funnel Stages')
plt.ylabel('Number of Users')
plt.title('User Funnel Analysis for Control and Test Groups')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

The Test Group consistently outperforms the Control Group in terms of website clicks, searches, and content views, indicating better user engagement.
However, as we move towards the final conversion stages (Add to Cart and Purchase), the gap between the groups narrows, suggesting that the Test Group's higher engagement in the initial stages doesn't necessarily translate to a proportionally higher number of purchases.

Busniess recommendations: 
1. Optimize the Post-Click Experience. The conversion rate (from 'Add to Cart' to 'Purchase') wasn't proportionally high, so review the wesite setting can be a good start for understanding the low rate.
2. Reduce Friction in the Checkout Process. Simplifying and streamlining the checkout process can lead to higher conversions.
3. Gather user feeback. Understaning what users wanted, and imporve the website to a user friendly site can definitely help the performance.

### Cost Efficiency

```python
# Calculating cost efficiency metrics for the control group
control_data['CPI'] = control_data['Spend [USD]'] / control_data['# of Impressions']
control_data['CPC'] = control_data['Spend [USD]'] / control_data['# of Website Clicks']
control_data['CPS'] = control_data['Spend [USD]'] / control_data['# of Searches']
control_data['CPVC'] = control_data['Spend [USD]'] / control_data['# of View Content']
control_data['CPAC'] = control_data['Spend [USD]'] / control_data['# of Add to Cart']

cost_efficiency_metrics = ['CPI', 'CPC', 'CPS', 'CPVC', 'CPAC']
control_cost_efficiency = control_data[cost_efficiency_metrics].mean()

control_cost_efficiency
```

```python
# Calculating cost efficiency metrics for the test group
test_data['CPI'] = test_data['Spend [USD]'] / test_data['# of Impressions']
test_data['CPC'] = test_data['Spend [USD]'] / test_data['# of Website Clicks']
test_data['CPS'] = test_data['Spend [USD]'] / test_data['# of Searches']
test_data['CPVC'] = test_data['Spend [USD]'] / test_data['# of View Content']
test_data['CPAC'] = test_data['Spend [USD]'] / test_data['# of Add to Cart']

cost_efficiency_metrics = ['CPI', 'CPC', 'CPS', 'CPVC', 'CPAC']
test_cost_efficiency = test_data[cost_efficiency_metrics].mean()

test_cost_efficiency
```

```python
# Data for plotting
control_efficiency_values = control_cost_efficiency.values
test_efficiency_values = test_cost_efficiency.values

# Creating the bar plot
plt.figure(figsize=(12, 8))
x = range(len(cost_efficiency_metrics))
width = 0.35  # width of the bars

# Control group bars
plt.bar([i - width/2 for i in x], control_efficiency_values, width=width, color='blue', label='Control Group')

# Test group bars
plt.bar([i + width/2 for i in x], test_efficiency_values, width=width, color='green', label='Test Group')

# Adding labels, title, and legend
plt.xlabel('Cost Efficiency Metrics')
plt.ylabel('Cost [USD]')
plt.title('Comparison of Cost Efficiency Metrics between Control and Test Groups')
plt.xticks(x, cost_efficiency_metrics, rotation=45)
plt.legend()
plt.grid(axis='y')

# Display the plot
plt.tight_layout()
plt.show()
```

As observed, the Test Group tends to be more cost-efficient in some metrics (like CPS and CPVC) but less so in others (like CPI and CPAC).

Busniess recommendations:
1. Re-evaluate Ad Spending:The Test Group has a notably higher Cost Per Impression (CPI) than the Control Group. This might indicate that the advertisements for the test campaign are being placed in higher-cost platforms or times. It's essential to assess if the increased spending is justified by the results.
2. Assess Ad Placement Strategies:The higher CPI for the Test Group suggests that the ad placements might be more expensive. Evaluate if these placements are resulting in the desired quality of user traffic. It might be worth exploring other channels or platforms that offer a better balance of cost and quality.
